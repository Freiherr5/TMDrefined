## mean_df

```python
all_annot = pd.read_excel("all_Annots_N_out.xlsx", "Preds")
all_annot
```

```python
all_annot.columns
```

```python
all_annot_drop = all_annot.dropna(subset="name")
```

```python
all_annot_drop
```

```python
use_mean = mean_df.reset_index()
window_mean_N = aaw.AAwindowrizer.get_aa_window_df(df=use_mean, column_id="entry", column_seq="sequence", window_size=4,
                                                     column_aa_position="start_pos_TMD", start_pos=True, start_set_positive=False).show_df()
window_mean_N_plus = aaw.AAwindowrizer.modify_label_by_ident_column(window_mean_N, all_annot_drop, "entry")
window_mean_N_plus
```

```python
window_mean_N_plus_new = window_mean_N_plus.show_df().dropna(how="any")
window_mean_N_plus_new_new = window_mean_N_plus_new.drop(window_mean_N_plus_new[window_mean_N_plus_new["window_right"]==""].index)
window_mean_N_plus_new_new
```

```python
# drop duplicates the smart way
pos_N = window_mean_N_plus_new_new[window_mean_N_plus_new_new["label"]==1][["window_right", "window_left"]]
neg_N = window_mean_N_plus_new_new[window_mean_N_plus_new_new["label"]==0][["window_right", "window_left"]]
neg_N
```

```python
for index_1, rows_1 in pos_N.iterrows():
    if rows_1.to_list() in neg_N.to_numpy().tolist():
        ind_N = neg_N.to_numpy().tolist().index(rows_1.to_list())
        neg_N = neg_N.drop(neg_N.index.tolist()[ind_N])
neg_N_new_index = neg_N.index.tolist()
```

```python
window_mean_N_plus_final = pd.concat([neg_N, pos_N]).drop_duplicates()
window_mean_N_plus_new_new_new = window_mean_N_plus_new_new.loc[window_mean_N_plus_final.index.tolist()].sample(frac=1)
window_mean_N_plus_new_new_new
```

```python
index_mean = window_mean_N_plus_new_new_new[window_mean_N_plus_new_new_new["label"]==1].sort_values("weights", ascending=False).index.tolist()[:100]
index_mean2 = window_mean_N_plus_new_new_new[window_mean_N_plus_new_new_new["label"]==0].sort_values("weights", ascending=False).index.tolist()[:600]
index_full = index_mean.extend(index_mean2)
mean_val = random.sample(index_mean, 200)

mean_val_df_N = window_mean_N_plus_new_new_new.loc[mean_val]
mean_val_df_N.to_excel("mean_val_df_N_balance0.xlsx")
mean_test_df_N = window_mean_N_plus_new_new_new.drop(mean_val)
mean_test_df_N.to_excel("mean_test_df_N_balance0.xlsx")
```
