# retrieving the necessary DataFrames

```python
import pandas as pd
import numpy as np
```

```python
all_df = pd.read_excel("all_Annots_N_out.xlsx", "Preds").set_index("entry")
all_df.head(10)
```

```python
membranome_df = all_df[all_df["annotation_software"]=="Membranome"].dropna(subset="tmd").drop(['len_signal', 'len_ecto', 'len_endo', 'top_prediction (Annot. AI / DB)'], axis=1)
membranome_df
```

```python
pm_membranome_df = membranome_df[membranome_df["subcellular_location"]=="Plasma memb."]
pm_membranome_df 
```

```python
index_mem = pm_membranome_df.index.tolist()
```

```python
uniprot_df = pd.read_excel("UniProt_SS_N_out.xlsx").set_index("entry")
uniprot_df.columns
```

```python
mem_seq_filter = uniprot_df.loc[index_mem, "sequence"]
```

```python
pm_mem_concat = pd.concat([pm_membranome_df, mem_seq_filter], axis=1)
pm_mem_concat
```

```python
expert_df = pd.read_excel("Expert_curated.xlsx").set_index("entry")
expert_df
index_exp= expert_df.index.tolist()
```

```python
list_for_concat = []
for (e_index, e_rows), (index, rows) in zip(expert_df[["tmd", "name"]].iterrows(), uniprot_df.loc[index_exp, ["sequence", "tmd"]].iterrows()):
    start_pos = rows[0].find(e_rows[0]) + 1
    stop_pos = rows[0].find(e_rows[0]) + len(e_rows[0])
    list_for_concat.append([e_index, start_pos, stop_pos])
data_concat = pd.DataFrame(list_for_concat, columns=["entry", "start", "stop"]).set_index("entry")
data_concat
```

```python
data_concat[data_concat["start"]==0]
```

```python
expert_filter = data_concat.drop(["P08887", "Q810F0"])
expert_filter
```

```python
new_ind_expert = expert_filter.index.tolist()
```

```python
uniprot_part_for_exp = uniprot_df.loc[new_ind_expert, ['name', 'gene_name', 'organism', 'dataset', 'sequence', 'len_sequence']]
```

```python
df_expert_finalized = pd.concat([expert_filter, uniprot_part_for_exp], axis=1)[['name', 'gene_name', 'organism', 'dataset','len_sequence', 'start', 'stop', 'sequence']]
df_expert_finalized
```

```python
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import pandas as pd
import numpy as np

N = 10
data = np.random.uniform(0, 45, size=(N, N))
for x, y in np.random.randint(0, N, 50).reshape(-1, 2):
    data[x, y] = np.nan  # fill in some nans at random places
df = pd.DataFrame(data)
up_triang = np.triu(np.ones_like(data)).astype(bool)
print(up_triang)
ax = sns.heatmap(df, cmap='flare', xticklabels=True, yticklabels=True, square=True,
                 linecolor='white', linewidths=0.5, mask = up_triang,
                 cbar=True, cbar_kws={'shrink': 0.6, 'pad': 0}, vmin=0, vmax=45)

ax.tick_params(left=False, bottom=False)
plt.show()
```
