# filter Data, create labels

```python
import pandas as pd
import numpy as np
from scripts.AA_window import AAwindowrizer as aaw
from scripts import StandardConfig

path, sep = StandardConfig.find_folderpath()
```

```python
allout_data = pd.read_excel("raw_data/all_Annots_N_out.xlsx", "Preds").set_index("entry")
```

```python
allout_data.head(10)
```

```python
get_sequences = pd.read_excel("raw_data/UniProt_all_Nout_filtered.xlsx").set_index("entry")[["name","sequence"]]
get_sequences.head(5)
```

```python
arithmetic_mean_all = allout_data[allout_data["annotation_software"]=="arithmetic mean"][["start_pos_TMD", "stop_pos_TMD"]]
arithmetic_mean_all.head(5)
```

```python
arithmetic_mean_seq = pd.concat([arithmetic_mean_all, get_sequences], axis=1).dropna()
arithmetic_mean_seq.head(5)
```

```python
list_pos_column = [{"pos_label" : "start_pos_TMD", "start_pos" : True}, {"pos_label" : "stop_pos_TMD", "start_pos" : False}]
for items in list_pos_column:
    arithmetic_mean_labels = aaw.get_aa_window_df(window_size=4, df=arithmetic_mean_seq, column_id="name", 
                                                  column_seq="sequence", column_aa_position=items["pos_label"], 
                                                  start_pos=items["start_pos"], start_set_positive=False)
    arithmetic_mean_labels_t2 = aaw.modify_label_by_ident_column(df_label=arithmetic_mean_labels.show_df(), df_compare=allout_data, column_id="name", threshold=2)
    print(arithmetic_mean_labels_t2.window_describe())
    arithmetic_mean_labels_t2.show_df().to_excel(f"labels{sep}am_t2_NO_BIAS_{items["pos_label"]}.xlsx")
```
